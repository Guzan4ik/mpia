## Подготовка файлов

````python
!python -m pip download -r requirements_hash.txt --require-hashes --no-clean -d libs
````

````python
libs_dir = './libs'
````

````python
import shutil
import zipfile
import os
from os import walk
````

````python
libs_archives = next(walk(libs_dir), (None, None, []))[2]
libs_archives
````

````python
for lib_archive in libs_archives:
    src = os.path.join(libs_dir, lib_archive)
    trg = os.path.join(libs_dir, lib_archive.replace('.whl', '').replace('.tar.gz', '').replace('.zip', '').replace('.tar', ''))
    if '.whl' in lib_archive:
        with zipfile.ZipFile(src, 'r') as zip_ref:
            zip_ref.extractall(trg)
    else:
        shutil.unpack_archive(src, trg)
````
